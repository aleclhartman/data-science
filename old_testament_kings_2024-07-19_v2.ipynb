```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for the kings of Israel and Judah
data = {
    'King': [
        # United Kingdom
        'Saul', 'David', 'Solomon', 
        # Israel (Northern Kingdom)
        'Jeroboam I', 'Nadab', 'Baasha', 'Elah', 'Zimri', 'Omri', 'Ahab', 'Ahaziah', 'Joram', 
        'Jehu', 'Jehoahaz', 'Jehoash', 'Jeroboam II', 'Zechariah', 'Shallum', 'Menahem', 
        'Pekahiah', 'Pekah', 'Hoshea', 
        # Judah (Southern Kingdom)
        'Rehoboam', 'Abijah', 'Asa', 'Jehoshaphat', 'Jehoram', 'Ahaziah', 'Athaliah', 
        'Joash', 'Amaziah', 'Uzziah', 'Jotham', 'Ahaz', 'Hezekiah', 'Manasseh', 'Amon', 
        'Josiah', 'Jehoahaz', 'Jehoiakim', 'Jehoiachin', 'Zedekiah'
    ],
    'Start Year': [
        # United Kingdom
        -1040, -1010, -970, 
        # Israel (Northern Kingdom)
        -930, -910, -909, -886, -885, -885, -874, -853, -852, -841, -814, -798, -793, 
        -753, -752, -752, -742, -740, -732, 
        # Judah (Southern Kingdom)
        -930, -913, -911, -870, -848, -841, -841, -835, -796, -767, -740, -732, -716, 
        -687, -643, -641, -609, -609, -598, -597
    ],
    'End Year': [
        # United Kingdom
        -1010, -970, -930, 
        # Israel (Northern Kingdom)
        -910, -909, -886, -885, -885, -874, -853, -852, -841, -814, -798, -793, -753, 
        -752, -752, -742, -740, -732, -722, 
        # Judah (Southern Kingdom)
        -913, -911, -870, -848, -841, -841, -835, -796, -767, -740, -732, -716, -687, 
        -643, -641, -609, -609, -598, -597, -586
    ],
    'Kingdom': [
        # United Kingdom
        'United Kingdom', 'United Kingdom', 'United Kingdom', 
        # Israel (Northern Kingdom)
        'Israel', 'Israel', 'Israel', 'Israel', 'Israel', 'Israel', 'Israel', 'Israel', 'Israel', 
        'Israel', 'Israel', 'Israel', 'Israel', 'Israel', 'Israel', 'Israel', 'Israel', 
        'Israel', 'Israel', 
        # Judah (Southern Kingdom)
        'Judah', 'Judah', 'Judah', 'Judah', 'Judah', 'Judah', 'Judah', 'Judah', 'Judah', 
        'Judah', 'Judah', 'Judah', 'Judah', 'Judah', 'Judah', 'Judah', 'Judah', 'Judah', 
        'Judah', 'Judah'
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert years to positive for plotting (BCE)
df['Start Year'] = -df['Start Year']
df['End Year'] = -df['End Year']

# Create the plot
fig, ax = plt.subplots(figsize=(12, 10))

# Plot each king's reign
for _, row in df.iterrows():
    ax.plot([row['Start Year'], row['End Year']], [row['King'], row['King']], 
            marker='o', linestyle='-', linewidth=4, markersize=8, 
            label=row['Kingdom'] if 'United' in row['Kingdom'] else None, 
            color='royalblue' if row['Kingdom'] == 'Israel' else 'forestgreen' if row['Kingdom'] == 'Judah' else 'gold')

# Adding annotations for better clarity
for i, row in df.iterrows():
    ax.annotate(row['King'], (row['Start Year'], i), xytext=(5, -5), textcoords='offset points', fontsize=10)
```

```python
# Revising the legend to correspond correctly to the colors used in the plot
import matplotlib.patches as mpatches

# Create the plot
fig, ax = plt.subplots(figsize=(12, 10))

# Plot each king's reign
for _, row in df.iterrows():
    color = 'royalblue' if row['Kingdom'] == 'Israel' else 'forestgreen' if row['Kingdom'] == 'Judah' else 'gold'
    ax.plot([row['Start Year'], row['End Year']], [row['King'], row['King']], 
            marker='o', linestyle='-', linewidth=4, markersize=8, color=color)

# Adding annotations for better clarity
for i, row in df.iterrows():
    ax.annotate(row['King'], (row['Start Year'], i), xytext=(5, -5), textcoords='offset points', fontsize=10)

# Formatting
ax.set_yticks(range(len(df)))
ax.set_yticklabels(df['King'])
ax.invert_yaxis()  # Invert y-axis to have the first king on top
ax.set_xlabel('Year (BCE)')
ax.set_title('Timeline of the Kings of Israel and Judah')

# Creating a custom legend
legend_patches = [
    mpatches.Patch(color='gold', label='United Kingdom'),
    mpatches.Patch(color='royalblue', label='Israel'),
    mpatches.Patch(color='forestgreen', label='Judah')
]
ax.legend(handles=legend_patches, loc='best')

# Display the plot
plt.tight_layout()
plt.grid(True)
plt.show()
```
